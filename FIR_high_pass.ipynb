```python
import numpy as np
import matplotlib.pyplot as plt


# Sample input signal (noisy)
n = np.arange(100)
x = np.sin(2 * np.pi * 0.05 * n) + 0.5 * np.random.randn(len(n))

# High-pass FIR kernel (derived from difference operation)
h_hp = np.array([1, -1])  # emphasizes rapid change (high freq)
y_hp = np.convolve(x, h_hp, mode='same')

plt.figure(figsize=(12, 4))
plt.plot(x, label='Original Signal')
plt.plot(y_hp, label='High-Pass Filtered', linewidth=2)
plt.title("FIR High-Pass Filter (Differencing)")
plt.xlabel("Sample Index")
plt.grid(True)
plt.legend()
plt.show()
```
