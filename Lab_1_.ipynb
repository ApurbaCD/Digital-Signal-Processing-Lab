```python
# Unit ramp
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(-10,10)
# n=np.linspace(1,10,10)

n=np.where(t>=0,t,0)

print(t)
print(n)

plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.axis([-5,10,0,10])
plt.plot(t,n)
plt.show()
```

```python
# Unit step
import sys
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-10,20)  #[-10, -9, -8, ..., 18, 19] 

unit_step=np.where(x>=0 , 1 , 0)
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-10,20,-1,2])

plt.stem(x,unit_step) #stem used for discrete value(discrete time signal) figure whether plot for continuous(continuous time signal)
plt.plot()
```

```python
# Exponential
import sys
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-10,20)  #[-10, -9, -8, ..., 18, 19] 

a = 0.8
x_exp = np.where(x >= 0, a**x, 0)
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-10,20,-1,2])

plt.stem(x,x_exp) #stem used for discrete value(discrete time signal) figure whether plot for continuous(continuous time signal)
plt.plot()


```
