```python
import numpy as np
import matplotlib.pyplot as plt

# Sample indices
n = np.arange(0, 20)

# Realization for fs = 200 Hz → ω = π/2
omega1 = np.pi / 2
x_200Hz = 3 * np.cos(omega1 * n)

# Realization for fs = 75 Hz → ω = 4π/3 (Aliasing)
omega2 = 4 * np.pi / 3
x_75Hz = 3 * np.cos(omega2 * n)

# Plotting both signals
plt.figure(figsize=(12, 5))

# 200 Hz Sampling
plt.subplot(1, 2, 1)
plt.stem(n, x_200Hz)
plt.title("Realized x[n] at fs = 200 Hz")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)

# 75 Hz Sampling (Aliased)
plt.subplot(1, 2, 2)
plt.stem(n, x_75Hz)
plt.title("Realized x[n] at fs = 75 Hz (Aliased)")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
