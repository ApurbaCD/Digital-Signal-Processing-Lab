```python
import numpy as np
import matplotlib.pyplot as plt

def correlation(x,y):
    len_x=len(x)
    len_y=len(y)
    len_r=len_x+len_y-1
    r=[0]*len_r
    # print(len_x)
    print(r)
    for i in range (len_r):
        for j in range (len_x):
            if(i-j>=0 and i-j<len_y):
                r[i]+=x[j]*y[i-j];
    return r;

    

xy=np.array([1,2,3,4])
xx=np.arange(0,4)

yy=np.array([1,2,3])
yx=np.arange(0,3)

plt.grid()
plt.xlabel('Time index')
plt.ylabel('Magnitude')
plt.axis([-1,4,-4,5])
plt.stem(xx,xy)
plt.show()

r_xy_y=correlation(xy,yy)
r_xy_x=np.arange(0,6)

print(r_xy_y)
plt.grid()
plt.xlabel('Time index')
plt.ylabel('Magnitude')
plt.axis([-1,7,0,20])
plt.stem(r_xy_x,r_xy_y)
plt.show()
```
