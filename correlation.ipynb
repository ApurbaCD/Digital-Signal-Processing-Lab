```python
import numpy as np
import matplotlib.pyplot as plt

# Your user-defined correlation function
def correlation(x, y):
    len_x = len(x)
    len_y = len(y)
    len_r = len_x + len_y - 1
    r = [0] * len_r
    y_pad = np.pad(y, (len_x - 1, len_x - 1), 'constant')
    
    for i in range(len_r):
        for j in range(len_x):
            r[i] += x[j] * y_pad[i + j]
    return r

# Input sequences
x = [1, 3, -2, 4]
y = [2, 3, -1, 3]
z = [2, -1, 4, -2]

# Compute correlations
r_xy = correlation(x, y)
r_yz = correlation(y, z)

# Lags
lags_xy = np.arange(-len(y)+1, len(x))
lags_yz = np.arange(-len(z)+1, len(y))

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(lags_xy, r_xy)
plt.title('Cross-Correlation: x[n] ⨂ y[n]')
plt.xlabel('Lag')
plt.ylabel('Correlation Value')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(lags_yz, r_yz)
plt.title('Cross-Correlation: y[n] ⨂ z[n]')
plt.xlabel('Lag')
plt.ylabel('Correlation Value')
plt.grid(True)

plt.tight_layout()
plt.show()

# Print values
print("r_xy =", r_xy)
print("r_yz =", r_yz)


# convolution x[n]∗h[n], correlation x[n]∗h[-n]
```
