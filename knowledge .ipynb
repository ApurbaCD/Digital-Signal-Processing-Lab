```python
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(0,12,5)   #creates a NumPy array starting at 1, going up to (but not including) 12, with a step size of 5.
                      #n = [1, 6, 11]
                      #Start = 1,Step = 5,Next values: 1, 6,11,16 (then 16, which is beyond 12 â†’ stop)

# n = np.arange(-10, 20) by default step size 1

plt.xlabel('X')
plt.ylabel("Y")
plt.axis([-5,5,-5,15]) #give the x axis start and end value ,y axis start and end value
plt.grid(True)
plt.plot(n)

# x = [0, 1, 2]

# y = [1, 6, 11]
```

```python
n = np.linspace(0, 10, 5) #n = [ 0,  2.5, 5, 7.5 ,10 ]   each difference=(end-start)/(Number_of_step-1)=(10-0)/(5-1)
print(n)
plt.xlabel('X')
plt.ylabel("Y")
plt.axis([-5,5,-5,15]) #give the x axis start and end value ,y axis start and end value
plt.grid(True)
plt.plot(n)
```

```python
x=np.arange(-10,20)  #[-10, -9, -8, ..., 18, 19] 

unit_step=np.where(x>=0 , 1 , 0)
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-10,20,-1,2])

plt.stem(x,unit_step) #stem used for discrete value(discrete time signal) figure whether plot for continuous(continuous time signal)
plt.plot()
```
