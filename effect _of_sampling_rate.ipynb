```python
# Consider the analog signal: xa(t)=3cos(200t)+5sin(600t)+10cos(1200t).
# Show the effect of sampling rate.

# You're working with a composite analog signal.

#     cos(200t)  frequency = 200/2=100 Hz

#     sin(600t)  frequency = 600/2=300 Hz

#     cos(1200t)  frequency = 1200/2=600 Hz

#  Maximum frequency = 600 Hz
#  Nyquist Rate = 2*600=1200 Hz

# So, to avoid aliasing, sampling frequency fs must be  1200 Hz.

# 1200 Hz (above Nyquist   no aliasing)

# 1000 Hz (below Nyquist   aliasing)

# 600 Hz (equal to max frequency   aliasing)


import numpy as np
import matplotlib.pyplot as plt

# Time for continuous signal
t_cont = np.linspace(0, 0.01, 1000)  # 10 ms range

# Continuous analog signal
x_cont = 3*np.cos(200*np.pi*t_cont) + 5*np.sin(600*np.pi*t_cont) + 10*np.cos(1200*np.pi*t_cont)

# Sampling rates
fs_list = [1200, 1000, 600]
titles = ["fs = 1200 Hz (No Aliasing)", "fs = 1000 Hz (Aliasing)", "fs = 600 Hz (Severe Aliasing)"]

plt.figure(figsize=(15, 8))

for i,fs in enumerate(fs_list):
    Ts = 1/fs
    t_samp = np.arange(0, 0.01, Ts)
    xa_samp = 3*np.cos(200*np.pi*t_samp) + 5*np.sin(600*np.pi*t_samp) + 10*np.cos(1200*np.pi*t_samp)

    plt.subplot(1, 3, i+1)
    plt.plot(t_cont, x_cont, 'k', label='Original Signal')
    plt.stem(t_samp, xa_samp, linefmt='r-', markerfmt='ro', basefmt=' ', label='Sampled Signal')
    plt.title(titles[i])
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()

```
