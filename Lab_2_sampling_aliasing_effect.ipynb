```python
import numpy as np
import matplotlib.pyplot as plt

#original continous time signal parameter
f_s=5
t_cont=np.linspace(0,1,1000)
x_cont=np.sin(2*np.pi*f_s*t_cont)
# plt.grid(True)
# plt.plot(t_cont,x_cont)

#sampling rate 

fs_high=20 # above Nyquist (no aliasing) (>=2*f_s)(20>2*5)
fs_low=8   #below Nyquist (aliasing occurs) (<=2*f_s) (8<2*5)

#sampled signal at high sampling frequency

t_high=np.arange(0,1,1/fs_high)  #1/fs_high is sampling time(that means after every 1/fs_high we sample the original signal)
x_high=np.sin(2*np.pi*f_s*t_high)

#sampled signal at low sampling frequency

t_low=np.arange(0,1,1/fs_low)
x_low=np.sin(2*np.pi*f_s*t_low)

# Plotting high

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.plot(t_cont,x_cont,'k',label='Original Signal')

# 'k' stands for black line. label='Original Signal' allows it to be included in the legend.

plt.stem(t_high,x_high,linefmt="b-",markerfmt="bo",basefmt=" ")

# Overlays the sampled version using stems (vertical lines from baseline).

# linefmt='b-': Blue lines from baseline to the sample.

# markerfmt='bo': Blue circles at sample points.

# basefmt=" ": Hides the horizontal base line (cleaner look).

plt.title(f'high Sampling Rate (fs={fs_high}Hz)')

# This uses an f-string to insert the value of fs_high dynamically into the title.

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# plt.legend() is used to display the legend on your plot, based on any labeled elements (like lines or stems)
# you've already added using the label= argument.

# Plotting low
plt.subplot(1,2,2)
plt.plot(t_cont,x_cont,'k',label='Original Signal')
plt.stem(t_low,x_low,linefmt='r-',markerfmt='ro',basefmt=" ")
plt.title(f'high Sampling Rate (fs={fs_low}Hz)->Alaising')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
```
