```python
# Consider the continuous-time analog signal x(t)=3cos(100t). Sample the analog
# signal at 200 Hz and 75 Hz. Show the discrete-time signal after sampling. 
# realization.

import numpy as np
import matplotlib.pyplot as plt

t_cont=np.linspace(0,0.1,1000)
# Frequency of signal f=100/2=50Hz
a=3
f_s=50
x_cont=a*np.cos(100*np.pi*t_cont)

f_high=200 #above Nyquist (no aliasing)
f_low=75  #below Nyquist (aliasing occurs)

t_high=np.arange(0,0.1,1/f_high)
x_high=a*np.cos(100*np.pi*t_high)

t_low=np.arange(0,0.1,1/f_low)
x_low=a*np.cos(100*np.pi*t_low)

#Ploting
plt.figure(figsize=(14,6))

# Plot 1: fs = 200 Hz
plt.subplot(1,2,1)
plt.plot(t_cont,x_cont,'k',label='Analog Signal')
plt.stem(t_high,x_high,linefmt='b-',markerfmt='bo',basefmt=' ',label='Sampled(200 Hz)')
plt.title("Sampling at 200 Hz (No Aliasing)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)


# Plot 2: fs = 75 Hz
plt.subplot(1,2,2)
plt.plot(t_cont,x_cont,'k',label='Analog Signal')
plt.stem(t_low,x_low,linefmt='r-',markerfmt='ro',basefmt=" ",label='Sampled (75 Hz)')
plt.title("Sampling at 75 Hz (Aliasing)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()



# Realization

# import numpy as np
# import matplotlib.pyplot as plt

# # Sample indices
# n = np.arange(0, 20)

# # Realization for fs = 200 Hz   = /2
# omega1 = np.pi / 2
# x_200Hz = 3 * np.cos(omega1 * n)

# # Realization for fs = 75 Hz   = 4/3 (Aliasing)
# omega2 = 4 * np.pi / 3
# x_75Hz = 3 * np.cos(omega2 * n)

# # Plotting both signals
# plt.figure(figsize=(12, 5))

# # 200 Hz Sampling
# plt.subplot(1, 2, 1)
# plt.stem(n, x_200Hz)
# plt.title("Realized x[n] at fs = 200 Hz")
# plt.xlabel("n")
# plt.ylabel("x[n]")
# plt.grid(True)

# # 75 Hz Sampling (Aliased)
# plt.subplot(1, 2, 2)
# plt.stem(n, x_75Hz)
# plt.title("Realized x[n] at fs = 75 Hz (Aliased)")
# plt.xlabel("n")
# plt.ylabel("x[n]")
# plt.grid(True)

# plt.tight_layout()
# plt.show()
```
