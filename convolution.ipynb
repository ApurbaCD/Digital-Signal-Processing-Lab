```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
def convol(x,h):
    len_x = len(x)
    len_h = len(h)
    len_y=len_x+len_h-1
    y=[0]*len_y;
    print(y)
    for i in range(len_y):
        for k in range(len_x):
            if(i-k>=0 and i-k<len_h):
                y[i]+=h[i-k]*x[k]

    return y

```
