```python
# highest rate of oscillation in a discrete-time sinusoidal is obtained
# when ω=π(pi)

# In discrete-time, a real sinusoid is given by:
#         x[n]=cos⁡(ωn)

# Angular frequency ω ranges from 0 to π (because of periodicity:
#        cos⁡(ωn) =cos⁡((ω+2π)n)

import numpy as np
import matplotlib.pyplot as plt

n=np.arange(0,20)

# ω = π
omega=np.pi

y=np.cos(omega*n)   # = (-1)^n


plt.stem(n,y)
plt.title(r'Discrete-Time Cosine with $\omega = \pi$: $\cos(\pi n) = (-1)^n$')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


```
