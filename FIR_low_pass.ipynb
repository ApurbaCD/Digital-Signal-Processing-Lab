```python
import numpy as np
import matplotlib.pyplot as plt

# Sample input signal (noisy)
n = np.arange(100)
x = np.sin(2 * np.pi * 0.05 * n) + 0.5 * np.random.randn(len(n))

# Low-pass FIR filter (moving average)
M = 5
h_lp = np.ones(M) / M
y_lp = np.convolve(x, h_lp, mode='same')

plt.figure(figsize=(12, 4))
plt.plot(x, label='Noisy Signal')
plt.plot(y_lp, label='Low-Pass Filtered', linewidth=2)
plt.title("FIR Low-Pass Filter (Moving Average)")
plt.xlabel("Sample Index")
plt.grid(True)
plt.legend()
plt.show()
```
